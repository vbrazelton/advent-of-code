```python
def normalize_num(num) -> int:
    if(num == "-"):
        return -1
    if(num == "="):
        return -2
    return int(num)

def snafu_to_decimal(number_str: str):
    result = 0
    for index, num in enumerate(reversed(number_str)):
        result += 5**index * normalize_num(num)
    return result

trans_rev = {
    4: '-',
    3: '=',
    2: '2',
    1: '1',
    0: '0'
}

def decimal_to_snafu(decimal: int) -> str:
    snafu = ''
    while decimal > 0:
        rem, decimal = decimal % 5, round(decimal / 5)
        snafu += trans_rev[rem]
    return snafu[::-1]
    
```

```python
decimal_to_snafu_answer_key = {
    "1": 1,
    2: "2",
    3: "1=",
    4: "1-",
    5: "10",
    6: "11",
    7: "12",
    8: "2=",
    9: "2-",
    10: "20",
    15: "1=0",
    20: "1-0",
    2022: "1=11-2",
    12345: "1-0---0",
    314159265: "1121-1110-1=0",
}

snafu_to_decimal_answer_key = {
    "1": 1,
    "2": 2,
    "1=": 3,
    "1-": 4,
    "10": 5,
    "11": 6,
    "12": 7,
    "2=": 8,
    "2-": 9,
    "20": 10,
    "1=0": 15,
    "1-0": 20,
    "1=11-2": 2022,
    "1-0---0": 12345,
    "1121-1110-1=0": 314159265,
}
```

```python
for x in snafu_to_decimal_answer_key.items():
    assert(snafu_to_decimal(x[0]) == x[1])
```
