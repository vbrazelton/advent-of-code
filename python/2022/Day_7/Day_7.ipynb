```python

class Line:
    def __init__(self, line: str):
        self.input_line = line
        self.line = None
        self.parse_line()

    def determine_type(self):
        if self.input_line[0] == '$':
            self.type = 'CMD'
            return
        if self.input_line[:3] == 'dir':
            self.type = 'DIR'
            return
        self.type = 'FILE'
        return
        

    def parse_line(self) -> bool:
        self.determine_type()

class FileSystem:
    def __init__(self):
        self.cw = []
        self.filetree = defaultdict(lambda: defaultdict(int))
        self.current_line = None
        self.lines_processed = 0
        self.continued = False

    def add_dir_to_filetree(self, dir):
        sub_dict = self.filetree
        # Iterate through the keys in fs.cw except for the last one
        for key in self.cw:
            sub_dict = sub_dict[key]
        # Use the last key in fs.cw to set the value
        sub_dict[dir] = defaultdict(lambda: defaultdict(int))

    def add_file_to_filetree(self, s):
        sub_dict = self.filetree
        # Iterate through the keys in fs.cw except for the last one
        for key in self.cw:
            sub_dict = sub_dict[key]
        # Use the last key in fs.cw to set the value
        sub_dict[s[1]] = int(s[0])

    def get_current_line(self):
        return self.current_line
    
    def increment_lines_processed(self):
        self.lines_processed += 1

    def set_current_line(self, line):
        self.increment_lines_processed()   
        self.current_line = Line(line)
        return
    
    def handle_CMD_CD(self, split_line: list[str]):
        if split_line[2] == '/':
            self.cw = ['/']
        elif split_line[2] == '..':
            self.cw.pop()
        else:
            self.cw.append(split_line[2])

    def handle_CMD_LS(self):
        return

    def handle_CMD(self):
        split_line = self.current_line.input_line.split()
        if split_line[1] == 'cd':
            self.handle_CMD_CD(split_line)
        elif split_line[1] == 'ls':
            self.handle_CMD_LS()
        else:
            Exception('Bad')

    def handle_DIR(self):
        split_line = self.current_line.input_line.split()
        self.add_dir_to_filetree(split_line[1])

    def handle_FILE(self):
        split_line = self.current_line.input_line.split()
        self.add_file_to_filetree(split_line)
    
    def process_line(self, line: str):
        self.set_current_line(line)
        if self.current_line.type == 'CMD':
            self.handle_CMD()
        if self.current_line.type == 'DIR':
            self.handle_DIR()
        if self.current_line.type == 'FILE':
            self.handle_FILE()

```
